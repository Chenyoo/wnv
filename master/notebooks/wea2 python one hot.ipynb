```python
wea2['CodeSum2']=wea2.apply(lambda row: row['CodeSum2'].split() , axis=1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
wea22 = wea2.join(pd.DataFrame(mlb.fit_transform(wea2.pop('CodeSum2')),
                          columns=mlb.classes_,
                          index=wea2.index))
```
