```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import numpy as np
from sklearn import metrics
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
## error = sum of squared errors (SSE)
def error(predicted, observed):
    err = sum(np.subtract(predicted, observed)**2)
    return(err)
```

```python
from sklearn.metrics import auc, roc_curve
```

```python
### We will create a table for comparing the results
res = pd.DataFrame(columns=['model','auc','accuracy','error'])
res = []
```

## XGBoost

```python
import sys
#!{sys.executable} -m pip install xgboost
import xgboost as xgb
```

```python
# Transform data to a DMatrix object
dtrain = xgb.DMatrix(np.array(X_train), label=y_train.values.ravel())
ddev = xgb.DMatrix(np.array(X_dev), y_dev.values.ravel())
dtest = xgb.DMatrix(np.array(X_test), y_test.values.ravel())
```

RUN 1 max_depth: 2, eta: 1,

```python
#1# Set the parameters for the xgboost
param = {'max_depth': 2, 'eta': 1, 'silent': 1, 'objective': 'binary:logistic'}
#param['nthread'] = 10
param['eval_metric'] = 'rmse'
```

```python
#2# Train the model using the training sets
num_round = 10
mod_xgboost = xgb.train(param, dtrain, num_round)
```

```python
#3# predictions using the TRAIN set

predT = mod_xgboost.predict(dtrain)

predT_class = np.where(predT >= 0.5, 1, 0)
pd.crosstab(y_train.values.ravel(), predT_class)
```

```python
#4# metrias of train
fpr, tpr, thresholds = metrics.roc_curve(y_train.values.ravel(), predT, pos_label=1)
auc1 = auc(fpr, tpr)
accT = accuracy_score(y_train.values.ravel(), predT_class)
#err = metrics.mean_squared_error(y_test, pred_class)
errT = error(y_train.values.ravel(), predT_class)
res.append({'model':'XGBoost TRAIN max_depth:2, eta:1','auc':auc1,'accuracy':accT,'err':errT})
res
```

```python
#5# Make predictions using the DEV set
predD = mod_xgboost.predict(ddev)

predD_class = np.where(predD >= 0.5, 1, 0)
pd.crosstab(y_dev.values.ravel(), predD_class)
```

```python
#6# metrias of DEV
fpr, tpr, thresholds = metrics.roc_curve(y_dev.values.ravel(), predD, pos_label=1)
auc1 = auc(fpr, tpr)
accD = accuracy_score(y_dev.values.ravel(), predD_class)
#err = metrics.mean_squared_error(y_dev.values.ravel(), pred_class)
errD = error(y_dev.values.ravel(), predD_class)
res.append({'model':'XGBoost DEV max_depth:2, eta:1','auc':auc1, 'accuracy':accD, 'err':errD})
res
```

```python
#7# Make predictions using the testing set
pred = mod_xgboost.predict(dtest)

pred_class = np.where(pred >= 0.5, 1, 0)
pd.crosstab(y_test.values.ravel(), pred_class)
```

```python
#8# metrias of TEST

fpr, tpr, thresholds = metrics.roc_curve(y_test.values.ravel(), pred, pos_label=1)
auc1 = auc(fpr, tpr)
acc = accuracy_score(y_test.values.ravel(), pred_class)
#err = metrics.mean_squared_error(y_test, pred_class)
err = error(y_test.values.ravel(), pred_class)
res.append({'model':'XGBoost TEST max_depth:2, eta:1','auc':auc1, 'accuracy':acc, 'err':err})
res
```
